```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your processed file
df = pd.read_csv('../data/processed/nepal_combined.csv')  # or korea_combined.csv
df['datetime'] = pd.to_datetime(df['datetime'])
df['month'] = df['datetime'].dt.month
df['year'] = df['datetime'].dt.year

# 🔥 1. Fire count per year
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='year')
plt.title("Fire Counts by Year")
plt.show()

# 🔥 2. Fire count by month (all years)
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='month', order=range(1,13))
plt.title("Fire Counts by Month")
plt.show()

# 💡 3. Brightness distribution
plt.figure(figsize=(6,4))
sns.histplot(df['brightness'], kde=True, bins=30)
plt.title("Brightness Distribution")
plt.show()

# 🎚️ 4. Confidence values
plt.figure(figsize=(6,4))
sns.countplot(x='confidence', data=df)
plt.title("Fire Confidence Levels")
plt.show()
```

```python
import pandas as pd
import plotly.express as px
```

```python
# Correct assignment
# Load the Korea and Nepal datasets using the correct relative path
korea_df = pd.read_csv('../data/processed/korea_combined.csv')
nepal_df = pd.read_csv('../data/processed/nepal_combined.csv')


# Add country labels
korea_df['country'] = 'Korea'
nepal_df['country'] = 'Nepal'
```

```python
# Combine both datasets
combined_df = pd.concat([nepal_df, korea_df], ignore_index=True)

# Drop missing/invalid coordinates
combined_df = combined_df.dropna(subset=['latitude', 'longitude'])
combined_df = combined_df[
    (combined_df['latitude'] >= -90) & (combined_df['latitude'] <= 90) &
    (combined_df['longitude'] >= -180) & (combined_df['longitude'] <= 180)
]
```

```python
fig = px.scatter_geo(
    combined_df,
    lat='latitude',
    lon='longitude',
    color='country',
    hover_name='datetime',  # Corrected column name
    title='NASA FIRMS Fire Locations: Nepal vs South Korea',
    projection='natural earth'
)

fig.update_layout(geo=dict(
    showland=True,
    landcolor='rgb(243, 243, 243)',
    showcountries=True,
    countrycolor='rgb(200, 200, 200)'
))

fig.show()
```
