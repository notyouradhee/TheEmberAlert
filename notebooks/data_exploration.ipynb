```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load data
df = pd.read_csv('../data/processed/nepal_combined.csv')

# Step 2: Combine date and time into datetime
df['datetime'] = pd.to_datetime(df['date'] + ' ' + df['time'])

# Step 3: Drop original columns if not needed
df = df.drop(columns=['date', 'time'])

# Step 4: Extract year and month for analysis
df['year'] = df['datetime'].dt.year
df['month'] = df['datetime'].dt.month

# Step 5: Plot fire counts by year
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='year')
plt.title("Fire Counts by Year")
plt.show()

# Step 6: Plot fire counts by month
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='month', order=range(1, 13))
plt.title("Fire Counts by Month")
plt.show()

# Step 7: Plot brightness distribution
plt.figure(figsize=(6,4))
sns.histplot(df['brightness'], bins=30, kde=True)
plt.title("Brightness Distribution")
plt.show()

# Step 8: Plot confidence distribution
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='confidence')
plt.title("Fire Confidence Levels")
plt.show()
```

```python
# Combine both datasets
combined_df = pd.concat([nepal_df, korea_df], ignore_index=True)

# Drop missing/invalid coordinates
combined_df = combined_df.dropna(subset=['latitude', 'longitude'])
combined_df = combined_df[
    (combined_df['latitude'] >= -90) & (combined_df['latitude'] <= 90) &
    (combined_df['longitude'] >= -180) & (combined_df['longitude'] <= 180)
]
```

```python
fig = px.scatter_geo(
    combined_df,
    lat='latitude',
    lon='longitude',
    color='country',
    hover_name='datetime',  # Corrected column name
    title='NASA FIRMS Fire Locations: Nepal vs South Korea',
    projection='natural earth'
)

fig.update_layout(geo=dict(
    showland=True,
    landcolor='rgb(243, 243, 243)',
    showcountries=True,
    countrycolor='rgb(200, 200, 200)'
))

fig.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load and prepare your data
df = pd.read_csv('../data/processed/nepal_combined.csv')
df['datetime'] = pd.to_datetime(df['date'] + ' ' + df['time'])  # other got datetime column but sorry mistakely got date and time different column for combined csv file :()

# 📍 Create regions based on latitude
bins = [25, 27, 29, 31]  # Adjust for Nepal's latitude range
labels = ['Southern', 'Central', 'Northern']
df['region'] = pd.cut(df['latitude'], bins=bins, labels=labels, include_lowest=True)

# 🔥 Count detections per region
region_counts = df['region'].value_counts().sort_values(ascending=False)

# 📊 Plot
plt.figure(figsize=(8,5))
sns.barplot(x=region_counts.index, y=region_counts.values, palette='Oranges')
plt.title("Wildfire Detections by Region (Latitude-based)")
plt.xlabel("Region")
plt.ylabel("Number of Detections")
plt.tight_layout()
plt.show()
```
